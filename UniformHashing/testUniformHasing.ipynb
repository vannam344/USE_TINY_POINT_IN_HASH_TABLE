```python
from UniformHashTable import UniformHashTable
from CollisionHandlers import UniformHashStrategy
import random

# Tạo bảng băm mô phỏng cache server
table = UniformHashTable(initial_size=256, strategy=UniformHashStrategy())

# Tạo 1000 user ngẫu nhiên
for i in range(1000):
    user_id = f"user_{i}"
    profile_data = {"score": random.randint(0, 1000)}
    table.insert(user_id, profile_data)

# Kiểm tra độ lấp đầy
used_slots = sum(1 for x in table.table if x is not None)
load_factor = used_slots / table.size

print(f"Stored {used_slots} profiles")
print(f"Load factor: {load_factor:.2f}")
```

```python
import matplotlib.pyplot as plt

positions = [i for i, v in enumerate(table.table) if v is not None]
plt.hist(positions, bins=20)
plt.title("Uniform distribution of user data across hash table slots")
plt.xlabel("Slot index")
plt.ylabel("Number of keys")
plt.show()
```
